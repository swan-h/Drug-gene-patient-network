```python
import pandas as pd
import pprint
drugbankdata = pd.read_csv("uniprot links.csv")  ##turning list into a dataframe using pandas library. 


drugbankdata.shape ##to find out the total number of rows and columns.
drugbankdata.iloc[1:5,3:4]  ##to explore the dataframe by specifying the row index, the column index --the last value is ignored.

drugbankdata["UniProt ID"] ##to extract a specific column by name. 
##to extract specific columns by name in the order specified, just add on the the above list by comma.

##calculating frequency of each duplicate rows using collection library.
import collections as cl
variable = drugbankdata["UniProt ID"]
cl.Counter(variable)

##finding the total number of duplicates in the column by specifying the column name. 
drugbankdata['is_duplicated'] = drugbankdata.duplicated(['UniProt ID'])
drugbankdata['is_duplicated'].sum()

```

```python
drugbankdata.groupby('UniProt ID').apply
df = drugbankdata[drugbankdata["UniProt ID"].duplicated()]
print(df)
```

```python
column = drugbankdata["UniProt ID"]

##using mygenee module to map uniprot id's to their corresponding gene symbols. 
import mygene as mge
mg = mge.MyGeneInfo()

##calling querymany method to specify parameters: input=uniprot, output=symbol, returning all duplicates and as a dataframe. 
out = mg.querymany(column, scopes='uniprot', fields='symbol', species='human', returnall=True, as_dataframe=True)

```
